```python
import random
```

```python
preguntas = [
    {
        "pregunta": "Â¿CuÃ¡l es la capital de Francia?",
        "opciones": ["A) Madrid", "B) ParÃ­s", "C) Roma", "D) BerlÃ­n"],
        "respuesta": "B"
    },
    {
        "pregunta": "Â¿CuÃ¡ntos planetas hay en el sistema solar?",
        "opciones": ["A) 7", "B) 8", "C) 9", "D) 10"],
        "respuesta": "B"
    },
    {
        "pregunta": "Â¿QuiÃ©n escribiÃ³ 'Don Quijote de la Mancha'?",
        "opciones": ["A) Lope de Vega", "B) Cervantes", "C) GÃ³ngora", "D) Quevedo"],
        "respuesta": "B"
    },
    {
        "pregunta": "Â¿En quÃ© aÃ±o llegÃ³ el hombre a la Luna?",
        "opciones": ["A) 1965", "B) 1967", "C) 1969", "D) 1971"],
        "respuesta": "C"
    },
    {
        "pregunta": "Â¿CuÃ¡l es el elemento quÃ­mico con sÃ­mbolo O?",
        "opciones": ["A) Oro", "B) OxÃ­geno", "C) Osmio", "D) Ozono"],
        "respuesta": "B"
    }
]

def jugar_trivial():
    print("\U0001F3B2 Bienvenido al juego Trivial \U0001F3B2\n")
    acertadas = 0
    contestadas = 0
    random.shuffle(preguntas)

    for i, p in enumerate(preguntas, 1):
        print(f"Pregunta {i}: {p['pregunta']}")
        for opcion in p["opciones"]:
            print(opcion)

        respuesta_usuario = input("Tu respuesta (A, B, C o D): ").upper()
        contestadas += 1
        if respuesta_usuario == "STOP":
            break

        if respuesta_usuario == p["respuesta"]:
            print("\u2705 Â¡Correcto!\n")
            acertadas += 1
        else:
            print(f"\u274C Incorrecto. La respuesta correcta era {p['respuesta']}.\n")
        print("La puntuaciÃ³n hasta el momento es:", acertadas)
        
    print("\U0001F389 Juego terminado \U0001F389")
    print(f"PuntuaciÃ³n final: {acertadas} / {contestadas}")
```

```python
jugar_trivial()
```
