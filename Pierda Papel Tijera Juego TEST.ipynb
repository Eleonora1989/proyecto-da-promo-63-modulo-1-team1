```python
opciones = ["piedra", "papel", "tijera"]

def decidir_ganador(jugador1, jugador2):
    if jugador1 == jugador2:
        return "empate"

    if (jugador1 == "piedra" and jugador2 == "tijera") or \
       (jugador1 == "papel" and jugador2 == "piedra") or \
       (jugador1 == "tijera" and jugador2 == "papel"):
        return "jugador1"

    return "jugador2"


def jugar():
    puntos_jugador1 = 0
    puntos_jugador2 = 0

    print("Bienvend@ al juego: \n\U0001FAA8 Piedra, \n\U0001F4C4 papel o \n\u2702\ufe0f tijera")
    print("El primero que llegue a 3 puntos gana.\n")

    while True:
        jugador1 = input("Jugador 1, Elige piedra/papel/tijera: ").strip().lower()
        jugador2 = input("Jugador 2, Elige piedra/papel/tijera: ").strip().lower()
    
        print(f"El jugador 1 eligi√≥: {jugador1}")
        print(f"El jugador 2 eligi√≥: {jugador2}")

        if jugador1 or jugador2 not in opciones:
            print("\u274C Opci√≥n no v√°lida. Usa piedra, papel o tijera.\n")
            continue


        resultado = decidir_ganador(jugador1, jugador2)

        if resultado == "empate":
            print("Empate!\n")
        elif resultado == "jugador 1":
            puntos_jugador1 += 1
            print("Jugador 1 gana esta ronda!\n")
        else:
            puntos_jugador2 += 1
            print("Jugador 2 gana esta ronda.\n")

        print(f"Marcador -> T√∫: {puntos_jugador1} | M√°quina: {puntos_jugador2}\n")

        # CONDICI√ìN DE FIN DE PARTIDA
        if puntos_jugador1 == 3:
            print("Felicidades Jugador 1 ¬°Has ganado la partida!")
            break

        if puntos_jugador2 == 3:
            print("Felicidades Jugador 2 ¬°Has ganado la partida!")
            break
```

```python
jugar()
```

```python
def piedra_papel_tijera():
    # Emojis definidos por Unicode
    PIEDRA = "\U0001FAA8"   # ü™®
    PAPEL  = "\U0001F4C4"   # üìÑ
    TIJERA = "\u2702\ufe0f" # ‚úÇÔ∏è

    opciones = {
        "piedra": PIEDRA,
        "papel": PAPEL,
        "tijera": TIJERA
    }

    reglas = {
        "piedra": "tijera",
        "tijera": "papel",
        "papel": "piedra"
    }

    j1 = input("Jugador 1 (piedra, papel o tijera): ").strip().lower()
    j2 = input("Jugador 2 (piedra, papel o tijera): ").strip().lower()

    if j1 not in opciones or j2 not in opciones:
        print("‚ùå Opci√≥n inv√°lida. Usa piedra, papel o tijera.")
        return

    print("\nElecciones:")
    print(f"Jugador 1: {opciones[j1]}")
    print(f"Jugador 2: {opciones[j2]}")

    if j1 == j2:
        print("\nü§ù Empate")
    elif reglas[j1] == j2:
        print("\nüèÜ Gana el Jugador 1")
    else:
        print("\nüèÜ Gana el Jugador 2")
```

```python
piedra_papel_tijera()
```
