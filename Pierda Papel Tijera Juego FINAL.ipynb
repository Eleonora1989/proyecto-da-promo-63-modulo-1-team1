```python
opciones = ["piedra", "papel", "tijera"]

def decidir_ganador(jugador1, jugador2):
    if jugador1 == jugador2:
        return "empate"

    if (jugador1 == "piedra" and jugador2 == "tijera") or \
       (jugador1 == "papel" and jugador2 == "piedra") or \
       (jugador1 == "tijera" and jugador2 == "papel"):
        return "jugador1"

    return "jugador2"


def jugar():
    puntos_jugador1 = 0
    puntos_jugador2 = 0

    print("Bienvend@ al juego: \n\U0001FAA8 Piedra, \n\U0001F4C4 papel o \n\u2702\ufe0f tijera")
    print("El primero que llegue a 3 puntos gana.")
    print("\u26A0 Si introduces una opciÃ³n no vÃ¡lida: el otro jugador gana la ronda.")
    print("\u26A0 Si ambos introducÃ­s una opciÃ³n no vÃ¡lida: nadie suma.\n")

    while True:
        jugador1 = input("Jugador 1, elige piedra/papel/tijera: ").strip().lower()
        

        jugador2 = input("Jugador 2, elige piedra/papel/tijera: ").strip().lower()
        
        # VALIDACIÃ“N (tu regla)
        valido1 = jugador1 in opciones
        valido2 = jugador2 in opciones
        print(f"El jugador 2 eligiÃ³: {jugador2}")
        print(f"El jugador 1 eligiÃ³: {jugador1}")
        
        if not valido1 and not valido2:
            print("\u274C \u274C Ambos jugadores introdujeron una opciÃ³n no vÃ¡lida. Nadie suma puntos.\n")
            continue

        if not valido1:
            puntos_jugador2 += 1
            print("\u274C Jugador 1 opciÃ³n no vÃ¡lida. Punto para Jugador 2.\n")
            print(f"Marcador -> Jugador 1: {puntos_jugador1} | Jugador 2: {puntos_jugador2}\n")
            continue

        if not valido2:
            puntos_jugador1 += 1
            print("\u274C Jugador 2 opciÃ³n no vÃ¡lida. Punto para Jugador 1.\n")
            print(f"Marcador -> Jugador 1: {puntos_jugador1} | Jugador 2: {puntos_jugador2}\n")
            continue

        # RONDA NORMAL (solo si ambos son vÃ¡lidos)
        resultado = decidir_ganador(jugador1, jugador2)

        if resultado == "empate":
            print("Â¡Empate!\n")
        elif resultado == "jugador1":
            puntos_jugador1 += 1
            print("Jugador 1 gana esta ronda!\n")
        else:
            puntos_jugador2 += 1
            print("Jugador 2 gana esta ronda!\n")

        print(f"Marcador -> Jugador 1: {puntos_jugador1} | Jugador 2: {puntos_jugador2}\n")

        # CONDICIÃ“N DE FIN DE PARTIDA
        if puntos_jugador1 == 3:
            print("Felicidades Jugador 1 \U0001F3C6 Â¡Has ganado la partida!")
            break

        if puntos_jugador2 == 3:
            print("Felicidades Jugador 2 \U0001F3C6 Â¡Has ganado la partida!")
            break
```
