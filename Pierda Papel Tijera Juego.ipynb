```python
import random
```

```python
opciones = ["piedra", "papel", "tijera"]

def decidir_ganador(jugador, maquina):
    if jugador == maquina:
        return "empate"

    if (jugador == "piedra" and maquina == "tijera") or \
       (jugador == "papel" and maquina == "piedra") or \
       (jugador == "tijera" and maquina == "papel"):
        return "jugador"

    return "maquina"


def jugar():
    puntos_jugador = 0
    puntos_maquina = 0

    print("üéÆ Piedra, papel o tijera (al mejor de 3)")
    print("El primero que llegue a 3 puntos gana.\n")

    while True:
        jugador = input("Elige piedra/papel/tijera: ").strip().lower()

        if jugador not in opciones:
            print("‚ùå Opci√≥n no v√°lida.\n")
            continue

        maquina = random.choice(opciones)
        print(f"ü§ñ La m√°quina eligi√≥: {maquina}")

        resultado = decidir_ganador(jugador, maquina)

        if resultado == "empate":
            print("‚û°Ô∏è Empate!\n")
        elif resultado == "jugador":
            puntos_jugador += 1
            print("‚úÖ Ganaste esta ronda!\n")
        else:
            puntos_maquina += 1
            print("‚ùå La m√°quina gana esta ronda.\n")

        print(f"Marcador -> T√∫: {puntos_jugador} | M√°quina: {puntos_maquina}\n")

        # üî¥ CONDICI√ìN DE FIN DE PARTIDA
        if puntos_jugador == 3:
            print("üèÜ ¬°Has ganado la partida!")
            break

        if puntos_maquina == 3:
            print("ü§ñ La m√°quina ha ganado la partida.")
            break
```
